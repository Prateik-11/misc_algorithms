```python
from math import sin
import matplotlib.pyplot as plt

def adams_bashforth(w_0, w_1, h, t_start, t_end, f):
    steps = int((t_end - t_start) / h)
    w_n_minus_1 = w_0
    w_n = w_1
    t_n_minus_1 = t_start
    t_n = t_n_minus_1 + h

    w_list = [w_0, w_1]
    t_list = [t_n_minus_1, t_n]

    for _ in range(2, steps):
        w_n_plus_1 = w_n + h * (1.5 * f(w_n, t_n) - 0.5 * f(w_n_minus_1, t_n_minus_1))
        w_list.append(w_n_plus_1)
        w_n_minus_1 = w_n
        w_n = w_n_plus_1
        t_n_minus_1 = t_n
        t_n = t_n + h
        t_list.append(t_n)

    return t_list, w_list
        
def f(x, t):
    return sin(t) - x
t_list, w_list = adams_bashforth(
    w_0 = 1,
    w_1= 0.9099979, 
    h = 0.1, 
    t_start = 0, 
    t_end = 10, 
    f = f)

plt.plot(t_list, w_list)
plt.title("w vs. t, Adams-Bashforth | h = 0.1 | t in [0, 10]")

```

```python
from math import sin
import matplotlib.pyplot as plt

def leapfrog(w_0, w_1, h, t_start, t_end, f):
    steps = int((t_end - t_start) / h)
    w_n_minus_1 = w_0
    w_n = w_1
    t_n_minus_1 = t_start
    t_n = t_n_minus_1 + h

    w_list = [w_0, w_1]
    t_list = [t_n_minus_1, t_n]

    for _ in range(2, steps):
        w_n_plus_1 = w_n_minus_1 + 2 * h * f(w_n, t_n)
        w_list.append(w_n_plus_1)
        w_n_minus_1 = w_n
        w_n = w_n_plus_1
        t_n_minus_1 = t_n
        t_n = t_n + h
        t_list.append(t_n)

    return t_list, w_list
        
t_list, w_list = leapfrog(w_0 = 1, 
w_1= 0.9099979, 
h = 0.1, 
t_start = 0, 
t_end = 10, 
f = f)

plt.plot(t_list, w_list)
plt.title("w vs. t, Leapfrog | h = 0.1 | t in [0, 10]")



```
