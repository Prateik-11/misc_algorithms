```python
import matplotlib.pyplot as plt

def question_4(w0, h, a, b):
    t = a
    w_t = w0
    t_list = []
    w_list = []
    while t <= b:
        print("t = " +str(t) + "    " +str(w_t))
        t_list.append(t)
        w_list.append(w_t)
        w_t = w_t + h + (h * ((t - w_t) ** 2))  
        t = t + h
    plt.plot(t_list, w_list)

question_4(w0 = 1, h = 0.005, a = 2, b = 6)
```

```python
import matplotlib.pyplot as plt

def f(t, y):
    return 1 + (t - y)**2

def question_6(w0, h, a, b):
    t = a
    w_t = w0
    t_list = []
    w_list = []
    while t <= b:
        print("t = " +str(t) + "    " +str(w_t))
        t_list.append(t)
        w_list.append(w_t)
        w_t = w_t + (h/2)*(f(t, w_t) + f(t + h, w_t + h*f(t, w_t)))
        t = t + h
    plt.plot(t_list, w_list)

question_6(w0 = 1, h = 0.005, a = 2, b = 6)
```
