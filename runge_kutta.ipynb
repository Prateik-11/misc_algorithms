```python
from math import sin, pi
import matplotlib.pyplot as plt

##############  SETTINGS ###############
steps = 400 # number of steps
h = (4 * pi) / steps # step size
t_0 = 0 # left boundary
x_0 = 1 # initial x value
w_0 = 0 # initial w value

def f_1(t_n, x_n, w_n):
    # derivative of x
    return w_n

def f_2(t_n, x_n, w_n):
    # derivative of w
    return - sin(x_n)

to_print = False # print output at each step
######################################

x_n = x_0
w_n = w_0
t_n = t_0

w_list = [w_0]
x_list = [x_0]
t_list = [t_0]

n = 0

if to_print:
    print(f"step {n} | At t = {t_n} : x = {x_n}, w = {w_n} ")

while n < steps:
    k_1_1 = h * f_1(t_n, x_n, w_n)
    k_1_2 = h * f_2(t_n, x_n, w_n)

    k_2_1 = h * f_1(t_n + h/2, x_n + (1/2)*k_1_1, w_n + (1/2)*k_1_2)
    k_2_2 = h * f_2(t_n + h/2, x_n + (1/2)*h*w_n, w_n - (h/2)*sin(x_n))

    k_3_1 = h * f_1(t_n + h/2, x_n + (1/2)*k_2_1, w_n + (1/2)*k_2_2)
    k_3_2 = h * f_2(t_n + h/2, x_n + (1/2)*k_2_1, w_n + (1/2)*k_2_2)

    k_4_1 = h * f_1(t_n + h, x_n + k_3_1, w_n + (1/2)*k_3_2)
    k_4_2 = h * f_2(t_n + h, x_n + k_3_1, w_n + (1/2)*k_3_2)

    x_n_plus_1 = x_n + (1/6) * (k_1_1 + 2 * k_2_1 + 2 * k_3_1 + k_4_1)  
    w_n_plus_1 = w_n + (1/6) * (k_1_2 + 2 * k_2_2 + 2 * k_3_2 + k_4_2)

    n = n + 1

    if to_print:
        print(f"step {n} | At t = {t_n} : x = {x_n}, w = {w_n} ")

    x_n = x_n_plus_1
    w_n = w_n_plus_1
    t_n = t_n + h

    w_list.append(w_n)
    x_list.append(x_n)
    t_list.append(t_n)

# Plot x vs w
plt.figure(0)
plt.clf()
plt.plot(x_list, w_list)
plt.title("Phase Plot: velocity w vs. x ")
plt.xlabel("x")
plt.ylabel("w")

# Plot t vs x
plt.figure(1)
plt.clf()
plt.plot(t_list, x_list)
plt.title("Position x vs. time ")
plt.xlabel("t")
plt.ylabel("x")
```
